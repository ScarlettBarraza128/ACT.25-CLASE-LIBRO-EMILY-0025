# Clase libro
* Crear la clase Libro, con 3 atributos (título, autor, editorial), 2 funciones leerdatos(), mostrardatos() y 2 objetos (libro1, libro2).

![texto del vínculo](https://raw.githubusercontent.com/ScarlettBarraza128/ACT.25-CLASE-LIBRO-EMILY-0025/refs/heads/main/infoclaselibroEmilyBarraza0025.png)

```python
class Libro:
    # 1. Metodo __init__ para inicializar atributos (constructor)
    def __init__(self, titulo, autor, editorial):
        # Atributos encapsulados
        self.titulo = titulo
        self.autor = autor
        self.editorial = editorial

    # 2. Comportamientos (métodos)
    def leerDatos(self):
        print(f"📖 Procesando información de '{self.titulo}'...")
        print(f"✅ Los datos han sido registrados correctamente en el sistema.")

    def mostrarDatos(self):
        print(f"📚 DETALLES DEL LIBRO:")
        print(f"   - Título: {self.titulo}")
        print(f"   - Autor: {self.autor}")
        print(f"   - Editorial: {self.editorial}")

# 3. Instanciar múltiples objetos con características únicas
libro1 = Libro("Cien años de soledad", "Gabriel García Márquez", "Editorial Sudamericana")
libro2 = Libro("Don Quijote de la Mancha", "Miguel de Cervantes", "Francisco de Robles")

# Llamadas a las funciones para cada objeto libro 1 y libro 2
print("🔖 Libro 1:")
libro1.leerDatos()
libro1.mostrarDatos()

print("\n🔖 Libro 2:")
libro2.leerDatos()
libro2.mostrarDatos()

print("\n✨ Creador de software por Emily Barraza NL = 7 ✨")
print("🏁 programa terminado")
```
